```python
from gurobipy import *

I,P =multidict({1:28,2:20,3:24,4:24})
J,d =multidict({1:22,2:14,3:26})
K,f =multidict({1:23,2:12,3:18,4:32})

c = {(1,1):10,(1,2):24,(1,3):36,
    (2,1):20,(2,2):16,(2,3):14,
    (3,1):30,(3,2):22,(3,3):12,
    (4,1):40,(4,2):30,(4,3):8}

model = Model("facility_location")

x = {}
y = {}

for i in I:
    for j in J:
        x[i,j] = model.addVar(vtype = "C", name = "x(%s,%s)"%(i,j))
for k in K:
    y[k] = model.addVar(vtype = "B", name = "y(%s)"%(k))

model.update()

for i in I:
    model.addConstr(quicksum(x[i,j]for j in J) <= P[i]*y[i])

for j in J:
    model.addConstr(quicksum(x[i,j] for i in I) == d[j])

model.setObjective(quicksum(f[k]*y[k] for k in K)+quicksum(c[i,j]*x[i,j] for (i,j) in x),GRB.MINIMIZE)

model.optimize()

print("Obj Value = ",model.ObjVal)

for v in model.getVars():
    if(v.X):
        print(v.VarName,v.X)
```
